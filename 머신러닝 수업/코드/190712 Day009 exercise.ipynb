# 1. 타이타닉 승객 데이터를 이용하여 다음 분석을 수행하시오. 

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

### 1) 세 개의 나이 그룹을 만든다.

```python
titanic.age.min()    # 0.42
titanic.age.max()    # 80.0
# 0~30: youth, 31~60: adult, 60~: senior
titanic["age_group"] = np.where(titanic["age"]>60, "Senior", np.where(titanic["age"]>30, "Adult", "Youth"))
titanic
```

### 2) 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산.

```python
titanic.pivot_table(
    index=["sex", "age_group"],
    columns="class",
    values="survived"
)
```

### 3) 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
titanic.pivot_table(
    index="sex",
    columns="class",
    values="survived"
)
```

### 4) 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고 각 그룹에 대해 '미성년자', '청년', '중년', '장년', '노년' 승객의 비율을 구한다.

```python
bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
```

```python
titanic["age_group2"] = pd.cut(titanic.age, bins, labels=labels)
titanic
```

```python
by_agegroup2 = titanic[titanic.age.notnull()].groupby(["survived", "age_group2"]).size()
by_agegroup2
```

```python
survival_size = titanic[titanic.age.notnull()].groupby("survived").size()
survival_size
```

```python
survival_ratio_by_agegroup2 = by_agegroup2 / survival_size
survival_ratio_by_agegroup2
```

```python
def arat(group, s):
    a = group["age_group2"].size()
    return a / s

titanic[titanic.age.notnull()].groupby(["survived", "age_group2"]).agg(arat, s=survival_size)
```

# 2. 아래의 작업은 임의의 데이터로 수행한다.

### 1) 회사의 상반기(1월~6월) 실적을 나타내는 데이터프레임과 하반기(7월~12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다.

```python
df1 = pd.DataFrame({
    "month": [1, 2, 3, 4, 5, 6],
    "매출": [5000, 2000, 3000, 6000, 7000, 8000],
    "비용": [3000, 1000, 4000, 4000, 3000, 2000]
})
df2 = pd.DataFrame({
    "month": [7, 8, 9, 10, 11, 12],
    "매출": [2000, 9000, 6000, 4000, 8000, 9000],
    "비용": [6000, 8000, 7000, 5000, 3000, 1000]
})
df1["이익"] = df1["매출"] - df1["비용"]
df2["이익"] = df2["매출"] - df2["비용"]
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

### 2) 1년간의 총 실적을 마지막 행으로 덧붙인다.

```python
df.loc[len(df)] = ["합계", df["매출"].sum(), df["비용"].sum(), df["이익"].sum()]
df
```
