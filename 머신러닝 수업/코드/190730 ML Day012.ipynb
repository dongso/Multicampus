# Multivariable Linear Regression (다중 선형 회귀)

```python
x1data = [70, 90, 88, 95, 75]    # 5명의 3월 모의고사 점수
x2data = [80, 88, 90, 98, 65]    # 5명의 4월 모의고사 점수
x3data = [75, 93, 90, 100, 70]    # 5명의 5월 모의고사 점수
ydata = [150, 185, 180, 200, 140]    # 5명의 실제 수능 점수
```

```python
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
w1 = tf.Variable(tf.random_normal([1]), name="weight1")
w2 = tf.Variable(tf.random_normal([1]), name="weight2")
w3 = tf.Variable(tf.random_normal([1]), name="weight3")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
hf = x1 * w1 + x2 * w2 + x3 * w3 + b
cost = tf.reduce_mean(tf.square(hf - y))
```

```python
train = tf.train.GradientDescentOptimizer(1e-5).minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cv, hv, _ = sess.run([cost, hf, train], feed_dict={x1:x1data, x2:x2data, x3:x3data, y:ydata})
    if step % 10 == 0:
        print("step: ", step, ", cost: ", cv, ", prediction: ", hv)

# 실제값: [150           185       180       200       140]
# 예측값: [148.3878  183.45149 180.90034 196.14665 147.95647]
```

## 행렬로 데이터 표현

```python
xdata = [[70, 90, 88, 95, 75], [80, 88, 90, 98, 65], [75, 93, 90, 100, 70]]    # 5명의 3, 4, 5월 모의고사 점수
ydata = [150, 185, 180, 200, 140]    # 5명의 실제 수능 점수
```

```python
np.shape(xdata)
```

```python
# (3, 5) -> (5, 3)
xdata = np.transpose(xdata)
np.shape(xdata)
```

```python
ydata = [[150], [185], [180], [200], [140]]
np.shape(ydata)
```

```python
x = tf.placeholder(tf.float32, shape=[None, 3])
y = tf.placeholder(tf.float32, shape=[None, 1])
w = tf.Variable(tf.random_normal([3, 1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
# hf = x * w + b  이렇게 쓰면 안됨
hf = tf.matmul(x, w) + b
cost = tf.reduce_mean(tf.square(hf - y))
train = tf.train.GradientDescentOptimizer(1e-5).minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cv, hv, _ = sess.run([cost, hf, train], feed_dict={x:xdata, y:ydata})
    if step % 10 == 0:
        print("step: ", step, ", cost: ", cv, ", prediction: ", hv)
```

# Queue Runner

```python
QUEUE_LENGTH = 20

# 큐 생성
q = tf.FIFOQueue(QUEUE_LENGTH, "float")

# 큐 러너 생성
enq_ops = q.enqueue_many(([1.0, 2.0, 3.0, 4.0],))
qr = tf.train.QueueRunner(q, [enq_ops, enq_ops, enq_ops])
```

```python
sess = tf.Session()
coord = tf.train.Coordinator()
threads = qr.create_threads(sess, coord=coord, start=True)

for step in range(10):
    print(sess.run(q.dequeue()))
    
coord.request_stop()    # 스레드 정지
coord.join(threads)    # 정지될 때까지 기다리는 함수
```

예제

```python
QUEUE_LENGTH = 20
q = tf.FIFOQueue(QUEUE_LENGTH, "float")
enq_ops1 = q.enqueue_many(([1.0, 2.0, 3.0, 4.0],))
enq_ops2 = q.enqueue_many(([5.0, 6.0, 7.0, 8.0],))
enq_ops3 = q.enqueue_many(([9.0, 10.0, 11.0, 12.0],))
qr = tf.train.QueueRunner(q, [enq_ops1, enq_ops2, enq_ops3])
```

```python
sess = tf.Session()
coord = tf.train.Coordinator()
threads = qr.create_threads(sess, coord=coord, start=True)

for step in range(30):
    print(sess.run(q.dequeue()))
    
coord.request_stop()
coord.join(threads)
```

파일에서 데이터 읽기 작업 순서
1. 파일 목록 읽기
2. 파일 목록을 파일이름큐에 저장
3. reader가 파일이름큐에서 파일명을 하나씩 읽어온다
4. decoder는 파일에 있는 데이터를 읽어들인다
5. 데이터를 정제(전처리)한다
6. 정제된 데이터를 example queue에 저장
7. 학습 진행

```python
queue = tf.train.string_input_producer(["q_1.txt", "q_2.txt", "q_3.txt"], shuffle=False)
coord = tf.train.Coordinator()
```

```python
reader = tf.TextLineReader()
key, value = reader.read(queue=queue)
print(sess.run(key))
print(sess.run(value))
```

```python
for i in range(100):
    speed, distance = tf.decode_csv(value, record_defaults=[[0], [0]])
    print(sess.run([speed, distance]))
coord.request_stop()
coord.join(threads)
```

# 연습문제 1

sqft_living에 따른 price?

(데이터(8:2) 분리)

```python
house = pd.read_csv("data/house2/train.csv")
house
```

```python
house1 = house[["price", "sqft_living"]]
house1
```

```python
xdata = list(house1["sqft_living"])
ydata = list(house1["price"])
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(xdata, ydata, test_size=0.2, random_state=42)
```

```python
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
w = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
hf = x * w + b
cost = tf.reduce_mean(tf.square(hf - y))
train = tf.train.GradientDescentOptimizer(1e-7).minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    sess.run(train, feed_dict={x:xtrain, y:ytrain})
    if step % 20 == 0:
        wv, cv, hv = sess.run([w, cost, hf], feed_dict={x:xtrain, y:ytrain})
        print("step: %s, w: %s, cost: %s, prediction: %s" % (step, wv, cv, hv))
```

```python
hvList = sess.run(hf, feed_dict={x:xtest})
```

```python
comparison = pd.DataFrame({
    "prediction": hvList,
    "ytest": ytest
})
comparison
```

(데이터(8:2) 분리)

```python
house2 = house[["price", "sqft_living", "bedrooms", "bathrooms"]]
house2
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(xdata, ydata, test_size=0.2, random_state=42)
```

```python
x = tf.placeholder(tf.float32, shape=[None, 3])
y = tf.placeholder(tf.float32, shape=[None, 1])
w = tf.Variable(tf.random_normal([3, 1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")

hf = x @ w + b
cost = tf.reduce_mean(tf.square(hf - y))
train = tf.train.GradientDescentOptimizer(1e-7).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
hvs = hvs.reshape(-1)
ytest = ytest.reshape(-1)
```
