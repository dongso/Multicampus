# 그룹화 연습

```python
years = range(1880, 2011)
pieces = []
for year in years:
    path = "data/babynames/yob%d.txt" % year
    frame = pd.read_csv(path, names=["name", "sex", "births"])
    frame["year"] = year
    pieces.append(frame)
names = pd.concat(pieces, ignore_index=True)
names
```

```python
total_births = names.pivot_table(
    index = "year",
    columns = "sex",
    values = "births",
    aggfunc = sum
)
total_births
```

## 특정 년도에 태어난 특정 성별의 아이들 중 특정 이름을 가진 아이들의 비율 구하기

```python
names.groupby(["year", "sex"]).births.sum()
# 그룹화 의미: 해당 년도 해당 성별의 전체 인원 수 구하기
```

```python
def add_prop(group):
    group["prop"] = group.births/group.births.sum()
    return group

names = names.groupby(["year", "sex"]).apply(add_prop)
names
```

## 연도별, 성별에 따른 빈도수가 가장 높은 이름 1000개 추출

```python
def get_top1000(group):
    return group.sort_values(by="births", ascending=False)[:1000]    # 그룹별로 ~1000행까지

grouped = names.groupby(["year", "sex"])
top1000 = grouped.apply(get_top1000)
top1000.index = np.arange(len(top1000))
top1000
```

```python
def get_top10(group):
    return group.sort_values(by="births", ascending=False)[:10]

grouped = names.groupby(["year", "sex"])
top10 = grouped.apply(get_top10)
top10
```

### 인기있는 이름 순으로 정렬했을 때 상위 n번째 이름이 전체 출생자 수에 대해 차지하는 비율을 구하려고 한다

```python
boys
```

```python
boys[boys.year==2010]
```

```python
boys[boys.year==2010].shape
```

```python
df = boys[boys.year==2010]
# df에 저장된 값을 중 prop을 보려고 한다
df.sort_values(by="prop", ascending=False)
```

```python
prop_cumsum = df.sort_values(by="prop", ascending=False).prop.cumsum()
prop_cumsum
```

* 누적합을 구했기 때문에 다음과 같은 사실을 알 수 있다. 예를 들어 인기있는 이름 상위 10번째까지는 전체 이름의 8.96%를 차지한다.

```python
prop_cumsum[:10]
```

```python
prop_cumsum.values.searchsorted(0.5)    # 정렬된 상태에서 0.5가 되는 위치를 찾는 것
```

* 즉, 출생아 비율 누적합이 0.5가 되는 지점은 상위 117(116+1)번째 이름이다.

```python
prop_cumsum.values[116]
```

```python
df = boys[boys.year==1900]
prop_cumsum = df.sort_values(by="prop", ascending=False).prop.cumsum()
prop_cumsum.values.searchsorted(0.5)    # 25 (24+1) 번째 까지가 상위 50%
prop_cumsum.values[24]
```

* 상위 50%를 차지하는 이름이 2010년에는 117개였던 반면 1900년에는 25개이다. 즉, 2010년에 비해 1900년에는 이름의 다양성이 부족했다

```python
df = boys[boys.year==1900]
in1900 = df.sort_values(by="prop", ascending=False).prop.cumsum()
in1900.values.searchsorted(0.5)+1    # 25 (24+1) 번째 까지가 상위 50%
```

### 그룹화 + apply 함수를 통해 위 작업을 자동화해서 다양성의 변화를 파악할 수 있다

```python
def get_quantile_count(group, q=0.5):    # quantile의 default 값은 0.5 -- 50% 지점을 찾는다
    group = group.sort_values(by="prop", ascending=False)
#     print(group)
    return group.prop.cumsum().values.searchsorted(q)
    
diversity = top1000.groupby(["year", "sex"]).apply(get_quantile_count)
diversity
```

```python
diversity.unstack("year")
```

```python
diversity.unstack("sex")
```

```python
diversity = diversity.unstack("sex")
diversity.plot(title="Number of Popular Names in Top 50%")
```

* 시간이 지날수록 상위 50%를 차지하는 이름의 개수가 많아지는 것을 (즉, 다양성이 높아지는 것을) 알 수 있다.

## json 열기 사용

```python
records = [json.loads(line) for line in open("data/bitly_usagov/example.txt", encoding="utf-8")]    # json.loads(line)이 리스트의 요소로 들어간다
records    # 모든 라인(딕셔너리)를 다 읽은 것
```

### 어느 시간대(tz)를 가장 많이 쓰는지 알고싶다

```python
[rec["tz"] for rec in records]    # 일부 딕셔너리에는 tz라는 column이 없기 때문에 에러가 난다
```

```python
time_zones = [rec["tz"] for rec in records if "tz" in rec]    # "rec 안에 tz 컬럼(key)이 있다면, 추출해라"
time_zones
```
