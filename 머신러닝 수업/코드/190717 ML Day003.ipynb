# CSS 선택자 추출

```python
html = """
<ul id="books">
<li id="b1">상실의 시대</li>
<li id="b2">데미안</li>
<li id="b3">이방인</li>
<li id="b4">태백산맥</li>
<li id="b5">어린 왕자</li>
</ul>
"""
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
sel = lambda q: print(soup.select_one(q).string)     # 람다 함수 이름은 sel, 전달되어지는 인자는 q
# 위 구문의 의미: sel 함수를 호출할 때 전달되어지는 인자(css 선택자)로부터 soup에 저장된 html문서에서 해당 텍스트를 추출하여 출력하겠다
```

```python
sel("ul > li#b4")
sel("ul li#b4")     # ul 태그 밑에 아이디가 b4인 li태그
sel("li#b4")
sel("#b4")    # b4라는 아이디가 유일하기 때문에 이렇게만 지정해줘도 된다
sel("#books #b4")
sel("#books > #b4")    # 아이디가 unique하다면 아이디만으로도 참조가 가능하다
```

```python
sel("li")    # 람다 함수 속 함수가 select_one인데 li태그는 여러개가 있으므로 처음 등장하는 하나만 출력된다
sel("li[id='b4']")    # 따옴표 안에 따옴표를 넣을때는 큰따옴표와 작은따옴표를 구분해서 써야 한다
sel("li:nth-child(2)")
sel("li:nth-of-type(2)")
```

# stackoverflow에서 질문과 답변 추출

```python
page = req.urlopen("https://stackoverflow.com/questions/tagged/python")
document = page.read()
soup = BeautifulSoup(document, "html.parser")
soup
```

```python
questions = soup.find(id="questions")
questions.find("a")
```

```python
questions.find_all("a")
```

```python
len(questions.find_all("a"))
```

```python
questions_list = questions.find_all("a", class_="question-hyperlink")
questions_list
```

```python
for questions in questions_list:
#     print(questions.string)
    print("질문: ", questions.get_text())
    print("=" * 40)
    print("답변: ", "http://stackoverflow.com"+questions.get("href"))
    print("=" * 40)
```
