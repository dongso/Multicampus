## 1. stack overflow (답변x=>'답변이 아직 없네요' 출력, 답변o(모든 답변 출력))

```python
def soup(url):
    return BeautifulSoup(req.urlopen(url).read(), "html.parser")
```

```python
home = soup("https://stackoverflow.com/questions/tagged/python")
home
```

```python
home_gist = home.select("div#questions a.question-hyperlink")
home_gist
```

```python
for link in home_gist:
    print("질문 제목: " + link.string)
    qpage = soup("https://stackoverflow.com" + link.attrs["href"])
    print("질문 내용: \n" + qpage.select("div.post-text")[0].text)
    print("질문 답변들: \n")
    ansList = qpage.select("div#answers div.answer div.post-layout div.answercell div.post-text")
    if ansList==[]:
        print("답변이 아직 없습니다.")
    else:
        for ans in ansList:
            print(ans.text)
    print("=" * 70)    
```

## 2. 특정 검색어에 해당되는 답변 글 긁어오기 (100 여개 정도...)

```python
def soup(url):
    return BeautifulSoup(req.urlopen(url).read(), "html.parser")
```

```python
# 검색어: 제헌절
rootaddr = "https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%EC%A0%9C%ED%97%8C%EC%A0%88&c_id=&c_name=&sm=tab_pge&kin_start="
```

```python
for i in range(1, 100, 10):
    spage = soup(rootaddr+str(i))
    qthumbs = spage.select("div.kinn ul.type01 dt.question a")
    for qthumb in qthumbs:
#         print("질문 제목: " + qthumb.text)
        qlink = qthumb.attrs["href"]
        qlink = qlink.encode('ascii', 'ignore').decode('ascii')
        qpage = soup(qlink)
        print("질문 제목: " + qpage.select("div.title")[0].text.strip())
        if qpage.select("div.c-heading__content") != []:
            print("질문 내용: " + qpage.select("div.c-heading__content")[0].text)
        else:
            print("질문 내용: (내용 없음)")
        if qpage.select("div._endContents.c-heading-answer__content > div") != []:
            for i in range(len(qpage.select("div._endContents.c-heading-answer__content > div"))):
                print("답변" + str(i+1) + ": " + qpage.select("div._endContents.c-heading-answer__content > div")[i].text)
        else:
            print("답변 내용: (답변이 아직 없음)")
        print("=" * 80)
    
```

## 3. 1부터 10,000까지 8이라는 숫자가 총 몇번 나오는가? 8이 포함되어 있는 숫자의 갯수를 카운팅 하는 것이 아니라 8이라는 숫자를 모두 카운팅 해야 한다. (※ 예를들어 8808은 3, 8888은 4로 카운팅 해야 함)

```python
startnum = 1
endnum = 10000
targetnum = 8
```

```python
cnttargets = 0
for i in range(startnum, endnum+1, 1):
    stringList = list(str(i))
    cnt = stringList.count(str(targetnum))
    cnttargets += cnt
print(cnttargets)
```
