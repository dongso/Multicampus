```
from keras.datasets import mnist
from keras import *
import numpy as np
from keras.utils import to_categorical
from konlpy.tag import Twitter
import nltk
from nltk.tokenize import word_tokenize
from konlpy.tag import Okt
from nltk.tokenize import WordPunctTokenizer
from konlpy.tag import Hannanum
from nltk.tokenize import *
from nltk.tag import *
from konlpy.tag import Kkma
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from collections import Counter
```

# keras를 이용한 CNN

```
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images.shape
test_labels
```

```
network = models.Sequential()
network.add(layers.Dense(512, activation="relu", input_shape=(28*28,)))    # 출력 갯수가 512, input_shape는 데이터 갯수를 마지막에 (그냥 비워두면 알아서 계산)
network.add(layers.Dense(10, activation="softmax"))
```

```
# 훈련 (컴파일)
network.compile(loss="categorical_crossentropy", optimizer="rmsprop", metrics=["accuracy"])
```

## 최적화: cost를 최소화하는 방향으로 파라미터를 업데이트 하는 것
1. 경사하강법(Gradient Descent): 기울기*학습률 => 가중치 갱신. 모델이 단순할 때는 괜찮은데 모델이 복잡할 때는 성능이 좋지 않다
2. 모멘텀: 운동량. 속도 크면 기울기를 크게 업데이트, 속도가 작으면 기울기를 작게 업데이트.
3. AdaGrad (RMSProp: 새로운 기울기만 학습률에 반영)
4. Adam: 모멘텀+AdaGrad


```
train_images = train_images.reshape((60000, 28*28))
test_images = test_images.reshape((10000, 28*28))
```

```
train_images = train_images.astype("float32")/255
test_images = test_images.astype("float32")/255
```

```
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```
test_cost, test_acc = network.evaluate(test_images, test_labels)
```

```
print(test_acc)
```
