# 자료형 변형

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.info()
```

* 자료형이 카테고리일때 뭔가 처리를 하려면 문자형으로 바꿔서 처리해야 한다

```python
tips["sex_str"] = tips["sex"].astype(str)
tips.info()
```

```python
tips["total_bill"] = tips["total_bill"].astype(str)
tips.info()
```

* 문자형으로 바꿨으므로 숫자연산은 못하고 문자열 연산만 할 수 있게 된다

```python
tips["total_bill"] = tips["total_bill"].astype(float)
tips.info()
```

## to_numeric

```python
tips_sub_miss = tips.head(10)
tips_sub_miss.loc[[1, 3, 5, 7], "total_bill"] = "missing"
tips_sub_miss
```

```python
tips_sub_miss.info()    # 문자열이 하나라도 있으면 그 열은 object
```

```python
tips_sub_miss["total_bill"].astype(float)
```

* 에러가 난다. "missing"이라는 단어를 어떻게 float으로 바꿔야 할지 모르기 때문에

```python
pd.to_numeric(tips_sub_miss["total_bill"])
```

* 역시 에러가 난다. 하지만 astype과 다른 점은 에러가 날 때 어떻게 할지 정하는 옵션이 있다는 점.

```python
pd.to_numeric(tips_sub_miss["total_bill"], errors="ignore")    # ignore: 아무 작업도 하지 않음
```

```python
pd.to_numeric(tips_sub_miss["total_bill"], errors="coerce")    # coerce: 숫자로 변환 안되는 유형의 데이터는 모두 NaN으로 처리
```

```python
pd.to_numeric(tips_sub_miss["total_bill"], errors="raise")    # raise: 에러를 그대로 발생시킴. default option.
```

```python
pd.to_numeric(tips_sub_miss["total_bill"], errors="coerce", downcast="float")
```

* downcast: 큰 자료형을 작은 자료형으로 바꿀 때 사용 (위 코드에서는 float64를 float32로 downcast)

## contour plot

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.contour.html

```python
def myFunc(x, y):
    return (1-x/2+x**4+y**2)*np.exp(-x**2-y**2)
x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
xx, yy = np.meshgrid(x,y)    # meshgrid: 등고선을 그리고자 하는 대상 데이터를 나타내는 특별(?)한 행렬. 지금 알 필요는 없다.
zz = myFunc(xx, yy)
zz
```

```python
plt.contourf(xx, yy, zz)
```

```python
plt.contourf(xx, yy, zz, alpha=.75)
```

```python
plt.contour(xx, yy, zz)
```

```python
plt.contourf(xx, yy, zz, alpha=.75, cmap="jet")
```
