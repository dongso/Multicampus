1. 여러 개의 파일(for문 사용)로드

```python
import glob

surveys = glob.glob("./data/survey_*")
print(surveys)
```

```python
import pandas as pd

list_surveys_df = []
for fp in surveys:
    df = pd.read_csv(fp)
    list_surveys_df.append(df)
print(len(list_surveys_df))
```

2. melt -> pivot_table 행/열 요소 변경 연습

```python
survey2 = list_surveys_df[2]
survey2
```

```python
survey2_melt = pd.melt(survey2, id_vars=["person", "quant"], var_name="status")
survey2_melt
```

```python
survey2_pt = survey2_melt.pivot_table(
    index=["person", "status"],
    columns="quant",
    values="value"
)
survey2_pt
```

3. ebola 데이터 조작

```python
ebola = pd.read_csv("data/country_timeseries.csv")
ebola
```

```python
ebola_melt = pd.melt(ebola, id_vars=["Date", "Day"])
ebola_melt
```

```python
variable_split = ebola_melt.variable.str.split("_")
variable_split
```

```python
ebola_melt["Status"] = variable_split.str.get(0)
ebola_melt["Country"] = variable_split.str.get(1)
ebola_melt
```
