# 차원 축소 - 주성분 분석 (PCA)

예제 1

```python
df = pd.DataFrame(columns=['calory', 'breakfast', 'lunch', 'dinner', 'exercise', 'body_shape'])
#칼로리, 식습관 운동습관과 그에 따른 체형 데이터를
df.loc[0] = [1200, 1, 0, 0, 2, 'Skinny']
df.loc[1] = [2800, 1, 1, 1, 1, 'Normal']
df.loc[2] = [3500, 2, 2, 1, 0, 'Fat']
df.loc[3] = [1400, 0, 1, 0, 3, 'Skinny']
df.loc[4] = [5000, 2, 2, 2, 0, 'Fat']
df.loc[5] = [1300, 0, 0, 1, 2, 'Skinny']
df.loc[6] = [3000, 1, 0, 1, 1, 'Normal']
df.loc[7] = [4000, 2, 2, 2, 0, 'Fat']
df.loc[8] = [2600, 0, 2, 0, 0, 'Normal']
df.loc[9] = [3000, 1, 2, 1, 1, 'Fat']
```

```python
df
```

```python
X = df[['calory', 'breakfast', 'lunch', 'dinner', 'exercise']]
Y = df[["body_shape"]]
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
x_std = StandardScaler().fit_transform(X)
x_std
```

```python
# 공분산행렬
# feature가 컬럼에 있는데, row로 바꾸기 위해 변환
features = x_std.T     # (5, 10)
features    # 각 열이 개인의 데이터를 의미
cov_matrix = np.cov(features)    # 공분산행렬 (5, 5)
cov_matrix
```

```python
# 공분산행렬로부터 아이겐벡터와 아이겐밸류 추출
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
print("고유벡터: %s" % eig_vecs)
print("고유값: %s" % eig_vals)
```

(일반적인) 고유벡터는 해당 행렬이 벡터에 작용하는 힘의 방향

공분산행렬의 고유벡터는 데이터가 어느 방향으로 분산되어 있는지를 나타내는 것

고유값? 고유벡터에 해당하는 상관계수

공분산행렬의 고유값은 각 축에 대한 공분산의 값

결국, 고유값이 큰 순서대로 고유 벡터를 나열하면, 중요한 주성분부터 추출이 된다.

```python
eig_vals[0] / sum(eig_vals)    # 0.73 => 73%의 정보를 표현할 수 있다
```

```python
project_X = x_std.dot(eig_vecs.T[0])
project_X    # 1차원으로 축소된 데이터
```

```python
res = pd.DataFrame(project_X, columns=["PC1"])
res["yaxis"] = 0.0
res["label"] = Y
res
```

```python
sns.lmplot("PC1", "yaxis", data=res, hue="label")    # 5차원 데이터가 1차원 데이터로 차원 축소
```

```python
df.head()
df_drop = df.drop(labels=["sales", "salary"], axis=1)
df_drop
```

```python
cols = df_drop.columns.tolist()
cols
```

```python
# "left" 컬럼의 위치 이동하기
cols.index("left")    # left의 위치
# cols.pop(cols.index("left"))    # 뽑아내기
cols.insert(0, cols.pop(cols.index("left")))    # 뽑아낸걸 0번 위치에 넣기
cols
```

```python
df_drop = df_drop.reindex(columns = cols)
df_drop
```

```python
X = df_drop.iloc[:, 1:8].values
Y = df_drop.iloc[:, 0].values
X.shape
Y.shape
```

```python
# 표준화
X_std = StandardScaler().fit_transform(X)
```

```python
# 공분산행렬 - 수식으로 구하기
mean_vec = np.mean(X_std, axis=0)
mean_vec
cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0] - 1)
cov_mat
```

```python
# 공분산행렬 - 함수 사용하기
np.cov(X_std.T)    # 행 인덱스가 컬럼에 와야 한다
# 위에 수식으로 구한 것과 동일하다
```

```python
sns.heatmap(cov_mat, annot=True, cmap="cubehelix", square=True)
```

```python
# eigenvalue, eigenvector 구하기
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print("아이겐벡터:\n%s" % eig_vecs)
print("아이겐밸류:\n%s" % eig_vals)
```

```python
eig_pairs = [(eig_vals[i], eig_vecs[:, i]) for i in range(len(eig_vals))]
eig_pairs
```

```python
eig_pairs.sort(key=lambda x: x[0], reverse=True)
eig_pairs
```

```python
for i in eig_pairs:
    print(i[0])    # eigenvalue들만 출력
```

```python
tot = sum(eig_vals)
var_exp = [(i/tot) * 100 for i in sorted(eig_vals, reverse=True)]
var_exp
```

PCA 클래스 사용하기

```python
pca = PCA().fit(X_std)
pca
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
# 원본 (7차원) 데이터가 6차원으로 축소되었을 때에 90% 이상이 보존된다
```

```python
sk_pca = PCA(n_components=6)    # 6차원으로 축소하고 싶다
y_sk = sk_pca.fit_transform(X_std)
y_sk
```
